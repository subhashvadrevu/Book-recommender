## Popularity Based Recommendation

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.shape
```

```python
num_rating_df = ratings_with_books.groupby(['Book-Title']).count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating': 'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_books.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns = {'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
merged_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
merged_df
```

```python
merged_df = merged_df[merged_df['num_ratings'] >= 250]
merged_df
```

```python
merged_df = merged_df.sort_values('avg_rating', ascending=False).reset_index()
merged_df
```

```python
merged_df = merged_df.merge(books, on='Book-Title').drop_duplicates('Book-Title').reset_index()
merged_df
```

```python
merged_df.drop(columns=['level_0'], inplace=True)
merged_df
```

```python
merged_df.shape
```

```python
final_ratings = filtered_rating_data[filtered_rating_data['Book-Title'].isin(famous_books)]
```

```python
final_ratings.duplicated().sum()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
np.where(pt.index=='1984')[0][0]
```

```python
sorted(list(enumerate(similarity_score[0])), key = lambda x : x[1], reverse=True)[1:6]
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    distances = similarity_score[index]
    similar_books = sorted(list(enumerate(similarity_score[index])), key = lambda x : x[1], reverse=True)[1:6]

    data = []
    for i in similar_books:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M']))

        data.append(item)
    return data
    # return suggestions
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Da Vinci Code')
```
